```python
expression_path='/home/yuhaowang/project/learning/BRCA-expression/COUNT_Symbol_matrix_transpose.csv'
data=pd.read_csv(expression_path,sep=',',index_col=0)
file_path='/home/yuhaowang/data/embedding/TCGA-BRCA'
#gigapath 验证所需要的dataset csv必须标签:slide id , 还有其他标签即可
file_list=os.listdir(file_path)
data['slide_id']=data.index.str[0:12]
```

```python
#TCGA-C8-A135 ->TCGA-C8-A135-01Z-00-DX1.h5  in  file list
for row in data.iterrows():
    slide_id=row[1]['slide_id']
    for file in file_list:
        if slide_id in file:
            data.loc[row[0],'slide_id']=file.replace('.h5','')
            break
```

```python
data.to_csv('/home/yuhaowang/project/learning/BRCA-expression/COUNT_Symbol_matrix_transpose_slide_id.csv',sep=',')
```
