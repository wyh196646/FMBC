```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# Example data
np.random.seed(42)
data = {
    "Training labels per class": np.repeat([1, 2, 4, 8, 16, 32], 20),
    "Few-shot performance": np.concatenate([
        np.random.normal(0.15, 0.05, 20),
        np.random.normal(0.2, 0.05, 20),
        np.random.normal(0.25, 0.05, 20),
        np.random.normal(0.3, 0.05, 20),
        np.random.normal(0.4, 0.05, 20),
        np.random.normal(0.45, 0.05, 20)
    ])
}

# Create a DataFrame
df = pd.DataFrame(data)

# Plot
plt.figure(figsize=(8, 6))
sns.boxplot(x="Training labels per class", y="Few-shot performance", data=df, palette="pastel")

# Customize the plot
plt.title("Few-shot Performance vs Training Labels per Class", fontsize=14)
plt.xlabel("Training labels per class", fontsize=12)
plt.ylabel("Few-shot performance", fontsize=12)
plt.ylim(0, 0.6)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```

```python

# 

def generate_ci_values(mean,low,high,runs=5):
 
    # 5
    numbers = np.linspace(low,high, 5)
    # n
    # 
    current_mean = np.mean(numbers)
    mean_difference = mean - current_mean
    # 0.676
    adjusted_numbers = numbers + mean_difference
    return adjusted_numbers
```

```python
metrics={
    "Balanced ACC": {
        "ResNet-50IN": (0.552, (0.490, 0.607)),
        "CTransPath": (0.639, (0.554, 0.728)),
        "REMEDIS": (0.676, (0.578, 0.770)),
        "UNI": (0.687, (0.617, 0.760))
    },
    "Weighted F1": {
        "ResNet-50IN": (0.516, (0.396, 0.629)),
        "CTransPath": (0.648, (0.536, 0.763)),
        "REMEDIS": (0.696, (0.591, 0.796)),
        "UNI": (0.691, (0.562, 0.800))
    },
    "AUROC": {
        "ResNet-50IN": (0.748, (0.675, 0.816)),
        "CTransPath": (0.840, (0.776, 0.903)),
        "REMEDIS": (0.864, (0.805, 0.915)),
        "UNI": (0.887, (0.833, 0.936))
    }
}

plot_data = []
for key, value in metrics.items():
    df=pd.DataFrame()
    for method, value_block in value.items():
        #print(method),add raw
        print(value_block)
        
        df[method] = generate_ci_values(value_block[0],value_block[1][0], value_block[1][1])
    plot_data.append(df)
```
